## handling null values

```python
imputer=SimpleImputer(strategy='mean')
imputer.fit(train_df[num_cols])
imputer.fit(test_df[num_cols])
```

```python
train_df[num_cols]=imputer.transform(train_df[num_cols])
test_df[num_cols]=imputer.transform(test_df[num_cols])
```

```python
train_df['HomePlanet']=train_df['HomePlanet'].fillna('Unknown')
test_df['HomePlanet']=test_df['HomePlanet'].fillna('Unknown')
```

```python
imputerx=SimpleImputer(strategy='most_frequent')
```

```python
imputerx.fit(train_df[cat_cols])
```

```python
train_df[cat_cols]=imputerx.transform(train_df[cat_cols])
test_df[cat_cols]=imputerx.transform(test_df[cat_cols])
```
